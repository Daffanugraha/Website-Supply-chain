```python
import json
import re
import math
import string

INPUT_FILE = "metadata_jurnal.json" 
OUTPUT_FILE = "data_preprocessed_chunk_32.json"

MAX_TOKENS = 32
OVERLAP_TOKENS = math.floor(MAX_TOKENS * 0.10)


def clean_text(text):
    if not text: return ""
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    pattern = f"[{re.escape(string.punctuation)}]"
    text = re.sub(pattern, ' ', text)
    return text

def case_folding(text):
    return text.lower().strip()

def create_chunks(text, size=64, overlap=6):
    words = text.split()
    chunks = []
    start = 0
    while start < len(words):
        end = start + size
        chunk = words[start:end]
        chunks.append(" ".join(chunk))
        if end >= len(words):
            break
        start += (size - overlap)
    return chunks


try:
    with open(INPUT_FILE, "r", encoding="utf-8") as f:
        raw_data = json.load(f)
except FileNotFoundError:
    print(f"File {INPUT_FILE} tidak ditemukan.")
    raw_data = []

meta_data_chunk_32 = []

if raw_data:
    for idx, item in enumerate(raw_data):
        doc_id = idx + 1
        abstract = item.get("abstract", "") or ""
        
        cleaned = clean_text(abstract)
        folded = case_folding(cleaned)
        chunks = create_chunks(folded, MAX_TOKENS, OVERLAP_TOKENS)
        

        meta_data_chunk_32.append({
            "id": doc_id,
            "chunks": chunks,
            "metadata": item
        })

        if idx == 0:
            print(f"ID DATA: {doc_id}")
            print("\n[DATA MENTAH]")
            print(abstract)
            
            print("\n[TEXT CLEANING]")
            print(cleaned)
            
            print("\n[CASE FOLDING]")
            print(folded)
            
            print("\n[RECURSIVE CHUNKING]")
            for i, c in enumerate(chunks):
                print(f"Chunk {i+1}: {c}")
            
            print("\n" + "="*50 + "\n")

    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        json.dump(meta_data_chunk_32, f, indent=4, ensure_ascii=False)

    print(f"Selesai! {len(meta_data_chunk_32)} data disimpan di {OUTPUT_FILE}")
```
